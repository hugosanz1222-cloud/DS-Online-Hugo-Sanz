```python
# Cargar el dataset
df = pd.read_csv("./data/california_cities.csv", index_col=0)

# Mostrar las primeras filas
print("Primeras filas del dataset:")
display(df.head())

# Descripción estadística
print("\nDescripción del dataset:")
display(df.describe())
```

```python
# Eliminar filas con valores nulos
df = df.dropna()

print(f"Filas después de eliminar nulos: {len(df)}")
display(df.head())
```

#### #1.2  

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center" o investigar como funciona bar_label). Los colores de las barras deben tener un 50% de transparencia y ser: azul, gris, rojo, aceituna y marrón.  

2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Arial", tamaño = 14, peso = "Bold", color azul acero o "steelblue")
3. El scatter de área de agua de un ciudad con su poblacion. Transparencia al 50% y color verde.
4. El boxplot de la altura de las ciudades.

```python
#Gráfico 1: Ciudades más pobladas
# Las 5 ciudades con más población
top5 = df.nlargest(5, 'population_total')[['city', 'population_total']]

fig1, ax1 = plt.subplots()
colores = ['blue', 'gray', 'red', 'olive', 'brown']
barras = ax1.bar(top5['city'], top5['population_total'] / 1000, color=colores, alpha=0.5)

# Añadir etiquetas de número sobre las barras
ax1.bar_label(barras, labels=[f"{v/1000:.3f}" for v in top5['population_total']], padding=3)

ax1.set_title('Ciudades más pobladas de California')
ax1.set_xlabel('Ciudades')
ax1.set_ylabel('Habitantes (miles)')

plt.tight_layout()
plt.show()
```

```python
#Gráfico 2: Histograma del área total
fig2, ax2 = plt.subplots()

ax2.hist(df['area_total_km2'], bins=50, color='steelblue', label='Hist Area')

ax2.set_title('Histograma del Área', fontname='Arial', fontsize=14, fontweight='bold')
ax2.set_xlabel(r'Superficie ($km^2$)', fontname='Arial', fontsize=14, fontweight='bold')
ax2.set_ylabel('Número de Ciudades', fontname='Arial', fontsize=14, fontweight='bold')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
#Gráfico 3: Scatter de área de agua vs población
fig3, ax3 = plt.subplots()

ax3.scatter(df['area_water_km2'], df['population_total'] / 1000,
            alpha=0.5, color='green')

# Encontrar outliers
outliers = df.nlargest(2, 'population_total')
ax3.annotate('Outliers',
             xy=(outliers.iloc[0]['area_water_km2'], outliers.iloc[0]['population_total'] / 1000),
             xytext=(60, 2000),
             arrowprops=dict(arrowstyle='->'))
ax3.annotate('',
             xy=(outliers.iloc[1]['area_water_km2'], outliers.iloc[1]['population_total'] / 1000),
             xytext=(60, 2000),
             arrowprops=dict(arrowstyle='->'))

ax3.set_title('Población por kilometro cuadrado de agua')
ax3.set_xlabel(r'Área de agua en la ciudad ($km^2$)')
ax3.set_ylabel('Población (miles habitantes)')

plt.tight_layout()
plt.show()
```

```python
#Gráfico 4: Boxplot de la altitud
fig4, ax4 = plt.subplots()

ax4.boxplot(df['elevation_m'], vert=False, patch_artist=True,
            boxprops=dict(facecolor='red'),
            flierprops=dict(marker='o', markerfacecolor='black', markersize=4))

ax4.set_title('Diagrama de caja de la Altitud de las ciudades de California')
ax4.set_xlabel('Altitud (m)')

plt.tight_layout()
plt.show()
```

```python
#Figura completa con los 4 gráficos
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
ax1, ax2, ax3, ax4 = axes[0, 0], axes[0, 1], axes[1, 0], axes[1, 1]

# Gráfico 1: Barras de ciudades más pobladas
top5 = df.nlargest(5, 'population_total')[['city', 'population_total']]
colores = ['blue', 'gray', 'red', 'olive', 'brown']
barras = ax1.bar(top5['city'], top5['population_total'] / 1000, color=colores, alpha=0.5)
ax1.bar_label(barras, labels=[f"{v/1000:.3f}" for v in top5['population_total']], padding=3)
ax1.set_title('Ciudades más pobladas de California')
ax1.set_xlabel('Ciudades')
ax1.set_ylabel('Habitantes (miles)')

# Gráfico 2: Histograma
ax2.hist(df['area_total_km2'], bins=50, color='steelblue', label='Hist Area')
ax2.set_title('Histograma del Área', fontname='Arial', fontsize=14, fontweight='bold')
ax2.set_xlabel(r'Superficie ($km^2$)', fontname='Arial', fontsize=14, fontweight='bold')
ax2.set_ylabel('Número de Ciudades', fontname='Arial', fontsize=14, fontweight='bold')
ax2.legend()

# Gráfico 3: Scatter
ax3.scatter(df['area_water_km2'], df['population_total'] / 1000, alpha=0.5, color='green')
outliers = df.nlargest(2, 'population_total')
ax3.annotate('Outliers',
             xy=(outliers.iloc[0]['area_water_km2'], outliers.iloc[0]['population_total'] / 1000),
             xytext=(60, 2000),
             arrowprops=dict(arrowstyle='->'))
ax3.annotate('',
             xy=(outliers.iloc[1]['area_water_km2'], outliers.iloc[1]['population_total'] / 1000),
             xytext=(60, 2000),
             arrowprops=dict(arrowstyle='->'))
ax3.set_title('Población por kilometro cuadrado de agua')
ax3.set_xlabel(r'Área de agua en la ciudad ($km^2$)')
ax3.set_ylabel('Población (miles habitantes)')

# Gráfico 4: Boxplot
ax4.boxplot(df['elevation_m'], vert=False, patch_artist=True,
            boxprops=dict(facecolor='red'),
            flierprops=dict(marker='o', markerfacecolor='black', markersize=4))
ax4.set_title('Diagrama de caja de la Altitud de las ciudades de California')
ax4.set_xlabel('Altitud (m)')

plt.tight_layout()
plt.show()
```
